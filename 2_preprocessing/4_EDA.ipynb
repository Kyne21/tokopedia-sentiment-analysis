#### Impute NaN values in appVersion & reviewCreatedVersion

```python
# Buat kolom tanggal saja untuk grouping, tanpa mengubah 'at'
df_analist['date'] = pd.to_datetime(df_analist['at']).dt.date

# Imputasi dengan mode per tanggal
for col in ['appVersion', 'reviewCreatedVersion']:
    mode_per_date = df_analist.groupby('date')[col].agg(lambda x: x.mode().iloc[0] if not x.mode().empty else np.nan)
    df_analist[col] = df_analist.apply(
        lambda row: mode_per_date[row['date']] if pd.isna(row[col]) else row[col], axis=1
    )

# Hapus kolom bantu 'date' jika tidak diperlukan
df_analist.drop(columns=['date'], inplace=True)
```

#### Drop anomalies in at & repliedAt

```python
# buatkan kode untuk mengecek jumlah entri data yang nilai kolom at > kolom repliedAt.
# Jika terdapat nilai NaN pada salah satu kolom untuk masing-masing entri, hitung jumlahnya juga secara terpisah

# Konversi kolom 'at' dan 'repliedAt' ke datetime
df_analist['at'] = pd.to_datetime(df_analist['at'])
df_analist['repliedAt'] = pd.to_datetime(df_analist['repliedAt'])

# Hitung jumlah entri di mana 'at' > 'repliedAt' (tidak termasuk NaN di kedua kolom)
count_at_greater_than_repliedAt = df_analist[(df_analist['at'].notna()) & (df_analist['repliedAt'].notna()) & (df_analist['at'] > df_analist['repliedAt'])].shape[0]

# Hitung jumlah entri dengan NaN di kolom 'at' atau 'repliedAt'
count_nan_in_dates = df_analist[(df_analist['at'].isna()) | (df_analist['repliedAt'].isna())].shape[0]

print(f"Jumlah entri dimana 'at' > 'repliedAt' (tanpa NaN): {count_at_greater_than_repliedAt}")
print(f"Jumlah entri dengan NaN di kolom 'at' atau 'repliedAt': {count_nan_in_dates}")
```

```python
# hapus entri data pada df_analist sejumlah nilai count_at_greater_than_repliedAt

# Identifikasi baris di mana 'at' > 'repliedAt' (tidak termasuk NaN)
indices_to_drop = df_analist[(df_analist['at'].notna()) & (df_analist['repliedAt'].notna()) & (df_analist['at'] > df_analist['repliedAt'])].index

# Hapus sejumlah entri data sesuai dengan count_at_greater_than_repliedAt
# Karena kita ingin menghapus *sejumlah* nilai yang sama, kita ambil slice dari index yang teridentifikasi
# Jika jumlah index_to_drop kurang dari count_at_greater_than_repliedAt, maka akan menghapus semua yang memenuhi kondisi
print(f"Jumlah entri awal di df_analist: {len(df_analist)}")
num_to_drop = min(len(indices_to_drop), count_at_greater_than_repliedAt)
df_analist.drop(indices_to_drop[:num_to_drop], inplace=True)

print(f"Jumlah entri yang dihapus: {num_to_drop}")
print(f"Jumlah entri tersisa di df_analist: {len(df_analist)}")
```

### Pertanyaan 1 - Berapa rata-rata waktu respon (selisih antara at dan repliedAt) dari pihak Tokopedia terhadap ulasan pengguna di Google Play Store Indonesia selama periode 1 Januari 2024 - 31 Desember 2024?

```python
# Filter data for the year 2024
df_2024 = df_analist[(df_analist['at'].dt.year == 2024)].copy()

# 1. Hapus row data yang tidak memiliki replyContent
df_responstime = df_2024.dropna(subset=['replyContent']).copy()

# 2. Hitung selisih sejak komen diposting dengan waktu reply dari pihak toped
df_responstime['response_time'] = df_responstime['repliedAt'] - df_responstime['at']

# 3. Buat 5 kategori waktu balas dan masukkan data ke kategori ini
def categorize_response_time(timedelta):
    if pd.isna(timedelta):
        return 'No Reply'
    hours = timedelta.total_seconds() / 3600
    if hours <= 1:
        return '0-1 hour'
    elif 1 < hours <= 6:
        return '1-6 hours'
    elif 6 < hours <= 12:
        return '6-12 hours'
    elif 12 < hours <= 24:
        return '12-24 hours'
    else:
        return '> 24 hours'

df_responstime['response_category'] = df_responstime['response_time'].apply(categorize_response_time)

# Display the results
display(df_responstime[['at', 'repliedAt', 'response_time', 'response_category']].head())
display(df_responstime['response_category'].value_counts())
```

```python
# Count the occurrences of each response category
response_category_counts = df_responstime['response_category'].value_counts()

# Sort the categories for better visualization in the new order
category_order = ['0-1 hour', '1-6 hours', '6-12 hours', '12-24 hours', '> 24 hours']
response_category_counts = response_category_counts.reindex(category_order)


# Create a bar chart
plt.figure(figsize=(10, 6))
response_category_counts.plot(kind='bar', color='skyblue')
plt.xlabel('Response Time Category')
plt.ylabel('Number of Reviews')
plt.title('Distribution of Tokopedia Response Times (Narrower Categories)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Pertanyaan 2 - Versi aplikasi Tokopedia mana yang memperoleh rata-rata skor ulasan tertinggi di Google Play Store Indonesia dengan hanya mempertimbangkan versi yang memiliki minimal 1000 ulasan agar hasil analisis representatif dan bebas bias dari jumlah sampel kecil selama periode 2023 - 2025?

```python
# Group by appVersion and count the number of reviews
app_version_counts = df_analist.groupby('appVersion').size()

# Filter app versions with at least 1000 reviews
filtered_app_versions = app_version_counts[app_version_counts >= 1000].index

# Filter the original DataFrame to include only the filtered app versions
df_filtered = df_analist[df_analist['appVersion'].isin(filtered_app_versions)].copy()

# Group the filtered DataFrame by appVersion and calculate the mean score
app_version_scores_filtered = df_filtered.groupby('appVersion')['score'].mean()

# Combine the average scores and counts into a single DataFrame
app_version_summary = pd.DataFrame({
    'average_score': app_version_scores_filtered,
    'review_count': app_version_counts[filtered_app_versions]
})

# Sort the results by app version in ascending order and take the top 10
top_10_app_versions_filtered = app_version_summary.sort_values(by='appVersion', ascending=True).head(10)

# Round the 'average_score' column to two decimal places
top_10_app_versions_filtered['average_score'] = top_10_app_versions_filtered['average_score'].round(2)


# Display the top 10 app versions with their average scores and review counts
print("Top 10 App Versions by Average Rating (with >= 1000 reviews):")
display(top_10_app_versions_filtered)
```

```python
import matplotlib.pyplot as plt

# Create a bar chart from the top_5_app_versions_filtered DataFrame
plt.figure(figsize=(10, 6))
plt.bar(top_10_app_versions_filtered.index, top_10_app_versions_filtered['average_score'], color='skyblue')
plt.xlabel('App Version')
plt.ylabel('Average Score')
plt.title('Top 5 App Versions by Average Rating (with >= 1000 reviews)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Pertanyaan 3 - Bagaimana perubahan rata-rata skor ulasan setiap bulan selama tahun 2025?

```python
# Analisis rata-rata skor ulasan per bulan selama 2025
df_2025 = df_analist[df_analist['at'].dt.year == 2025].copy()

if df_2025.empty:
    print('Tidak ada data ulasan untuk tahun 2025.')
else:
    df_2025['year_month'] = df_2025['at'].dt.to_period('M')
    monthly_avg_score_2025 = (
        df_2025.groupby('year_month')['score']
        .mean()
        .reset_index()
        .sort_values('year_month')
        .rename(columns={'score': 'avg_score'})
    )
    monthly_avg_score_2025['year_month'] = monthly_avg_score_2025['year_month'].dt.to_timestamp()

    display(monthly_avg_score_2025)

    plt.figure(figsize=(12, 6))
    plt.plot(monthly_avg_score_2025['year_month'], monthly_avg_score_2025['avg_score'], marker='o')
    plt.title('Perubahan Rata-rata Skor Ulasan per Bulan (2025)')
    plt.xlabel('Bulan (2025)')
    plt.ylabel('Rata-rata Skor')
    plt.xticks(rotation=45)
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()

```

Rata-rata skor ulasan Tokopedia sepanjang 2025 terus melemah: setelah sempat menanjak dari 2,2 di Januari ke puncak sekitar 2,36 di Februari, nilainya menurun stabil hingga berada di kisaran 1,65 pada November. Tren penurunan konsisten sejak Maret menunjukkan bahwa persepsi pengguna terhadap aplikasi makin negatif tiap bulan, hanya ada jeda kenaikan kecil di Oktober sebelum jatuh lagi.
