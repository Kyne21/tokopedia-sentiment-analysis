## LABELING

### Drop Neutral

karena netral susah dijelaskan kenetralannya dan beberapa algoritma labeling susah untuk dipertanyakan kevalidasiannya. maka kita akan menggunakan score untuk menentukan labelnya dan menghapus score 3 (netral) 

```python
# Create copy of the dataframe to avoid modifying the original
df_labeled = hasil_cleaning.copy()

# Count score 3 before dropping
neutral_count = len(df_labeled[df_labeled['score'] == 3])
print(f"Total data dengan score 3 (neutral) yang akan di-drop: {neutral_count}")

# Drop rows where score is 3
df_labeled = df_labeled[df_labeled['score'] != 3]

# Create labels based on score
df_labeled['label'] = df_labeled['score'].apply(lambda x: 'positive' if x > 3 else 'negative')

# Show distribution of labels
print("\nDistribusi label:")
print(df_labeled['label'].value_counts())

# Update hasil_cleaning with the cleaned version
hasil_cleaning = df_labeled
```
