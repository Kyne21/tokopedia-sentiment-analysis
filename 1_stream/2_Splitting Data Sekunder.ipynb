```python
import pandas as pd
import numpy as np
```

## Data Stream to CSV

```python
Stream1 = pd.read_csv('Scrape1.csv')
Stream2= pd.read_csv('Scrape2.csv')
```

```python
Data_Stream = pd.concat([Stream1, Stream2], ignore_index = True)
Data_Stream.info()
```

```python
Data_Stream.duplicated().sum()
```

```python
Data_Stream['Tanggal'] = pd.to_datetime(Data_Stream['at'])
Data_Stream['Tahun'] = Data_Stream['Tanggal'].dt.year

Data_Stream = Data_Stream.sort_values(by='Tanggal')
Data_Stream.head()
```

```python
Data_Stream = Data_Stream.sort_values(by='Tanggal', ascending=False)
Data_Stream.head()
```

```python
Data_Stream.to_csv('Scraping_Combined.csv', index = False)
```

## Data Sekunder

```python
Sekunder = pd.read_csv('data_sekunder.csv')
Sekunder.info()
```

```python
Sekunder['Tanggal'] = pd.to_datetime(Sekunder['at'])
Sekunder['Tahun'] = Sekunder['Tanggal'].dt.year
Sekunder['Tahun'].unique()
```

```python
Sekunder['Tahun'].value_counts()
```

```python
# Ambil 4946 data tahun 2024
sekunder_2024 = Sekunder[Sekunder['Tahun'] == 2024].sample(n=4946)

# Ambil 26054 data tahun 2023 dengan filter tanggal <= 2023-12-31
sekunder_2023 = Sekunder[(Sekunder['Tahun'] == 2023) & (Sekunder['Tanggal'] <= '2023-12-31')].sample(n=26054)

# Gabungkan menjadi satu dataframe baru
Sekunder_31k = pd.concat([sekunder_2024, sekunder_2023]).sort_values(by='Tanggal', ascending=False)
```
